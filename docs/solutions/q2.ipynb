# Minimax Algorithm

Minimax is a decision-making algorithm used in two-player, zero-sum games such as:

- Chess
- Tic-Tac-Toe
- Pacman (multi-agent version in this project)

## Goal:

- MAX player: maximize score
- MIN player: minimize score

## Core Idea

Minimax assumes that the opponent always plays optimally.

And so:
- MAX chooses the best possible outcome
- MIN chooses the worst possible outcome for MAX

## Example

```python
     MAX
    /    \
  MIN    MIN
 /   \   /   \
3     5 2     9
```


Step 1: MIN layer
- min(3, 5) = 3
- min(2, 9) = 2

Step 2: MAX layer
- max(3, 2) = 3

Final decision for MAX = 3

## Formal Definition

Let:
- S = state
- A = actions
- successor(S, a) = next state

### Terminal state:
V(S) = Evaluation(S)

### MAX node:
V(S) = max V(S')

### MIN node:
V(S) = min V(S')

## How to solve this problem with Minimax Algorithm?

Differences from normal minimax:

- Multiple agents (Pacman + ghosts)
- Pacman = MAX
- Ghosts = MIN
- Depth increases only when Pacman moves
- Agents rotate in order

## Complexity

Time complexity: $O(b^d)$

Where:
- $b$ = branching factor
- $d$ = depth

Problem:
- exponential growth
- very slow for deep search trees

## Limitations

- Very inefficient
- Not practical for deep games
- Needs optimization (Alpha-Beta Pruning)

## Applications

- Deep Blue (Chess AI, 1997)
