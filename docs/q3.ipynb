# Alpha-Beta Pruning

Alpha-Beta Pruning is an optimization technique for the Minimax algorithm.

It produces the exact same result as Minimax, but reduces the number of nodes explored during the search.

Hence makes the algorithm significantly faster.

## Goal

(Similar to Minimax), BUT:
- MAX player: maximize score
- MIN player: minimize score
- Skip branches that cannot affect the final decision (Key difference)


## Core Idea

While searching the game tree, Alpha-Beta keeps track of:

- **Alpha ($\alpha$):** the best value MAX can guarantee so far
- **Beta ($\beta$):** the best value MIN can guarantee so far

If at any point: $\alpha > \beta$

then the current branch can no longer influence the final decision and can be safely ignored.

This process is called **pruning**.

## Alpha and Beta

### Alpha ($\alpha$)

The highest score MAX has found so far.

Initial value:

```python
alpha = float('-inf')
```

### Beta ($\beta$)

The lowest score MIN has found so far.

Initial value:

```python
beta = float('inf')
```

## Example

### Step 1: Left branch

MIN chooses:

$\min(3, 5)=3$

MAX now has: $\alpha = 3$

### Step 2: Right branch

MIN examines: 2

So: $\beta = 2$

Now: $\alpha = 3$, $\beta = 2$

Since: $\alpha \ge \beta$

There is no point of exploring the next branch if it is guaranteed that MAX will choose the left branch value.

Therefore, the remaining branch does not need to be explored.

The final node is pruned.

## Why Pruning Works

If MAX already has a better option elsewhere, then:

- MAX will never choose a worse branch
- Exploring the remaining nodes becomes useless

Therefore:
- the result will not change
- unnecessary computation can be skipped

## Formal Definition

### MAX node

```python
value = float('-inf')

for child in children:
    value = max(value, child_value)

    alpha = max(alpha, value)

    if alpha >= beta:
        break
```

### MIN node

```python
value = float('inf')

for child in children:
    value = min(value, child_value)

    beta = min(beta, value)

    if alpha >= beta:
        break
```

## How Alpha-Beta Works in This Project

Differences from standard Alpha-Beta:

- Multiple agents (Pacman + ghosts)
- Pacman = MAX
- Ghosts = MIN
- Depth increases only when Pacman moves
- Agents rotate in order
- Berkeley implementation prunes only when:

```python
alpha > beta
```

instead of:

```python
alpha >= beta
```

## Complexity

Worst case:

$O(b^d)$

Best case with optimal pruning:

$O(b^{d/2})$

Where:
- $b$ = branching factor
- $d$ = depth

This allows much deeper searches compared to standard Minimax.

## Advantages

- Much faster than Minimax
- Reduces unnecessary exploration
- Produces the same optimal move
- Makes deep search practical

## Limitations

- Performance depends heavily on move ordering
- Worst-case complexity is still exponential
- More difficult to debug than standard Minimax
